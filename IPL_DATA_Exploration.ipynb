#FRANCHISE ANALYSIS


```python
grp=ipl.groupby('franchise')['final price (in lacs)'].sum()
grp
```

```python
grp.index.values
```

```python
plt.bar(grp.index, grp.values)
plt.xticks(rotation=90) # Rotate x-axis labels for better readability
plt.xlabel('Franchise') # Add x-axis label
plt.ylabel('Total Final Price (in lacs)') # Add y-axis label
plt.title('Total Final Price Spent by Each Franchise') # Add title
plt.show()
```

```python
plt.pie(grp.values, labels=grp.index,autopct='%1.2f',explode=[0,0,0,0,0,0,0,0,0,.3],startangle=90,shadow=True)
plt.show()
```

```python
import pandas as pd

# Assuming df is your DataFrame containing data about statuses and franchises

# Grouping by 'status' and 'franchise', and counting occurrences
st = ipl.groupby('status')['franchise'].value_counts()

# Selecting counts for 'RETAINED' status
st1 = st.loc['RETAINED']

# Finding the maximum count of franchises for 'RETAINED' status
max_count = st1.max()

print("Maximum count of franchises for 'RETAINED' status:", max_count)
```
